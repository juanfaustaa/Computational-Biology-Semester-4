# Import file

```python
from Bio import SeqIO
```

```python
seqA = SeqIO.read("fasta1.fasta", 'fasta')
seqB = SeqIO.read('fasta2.fasta', 'fasta')
seqZ = SeqIO.read('genbank1.genbank', 'genbank')

print(seqZ.seq)
sequenceA = seqA.seq
sequenceB = seqB.seq

print(sequenceA)
print(sequenceB)
```

# sequence manipulation using python

```python
#a
seqArev = seqA.seq[::-1]
print(seqArev, '\n')

#b
complimentseqB = seqB.seq.complement()
print(complimentseqB, '\n')

#c 
seqC = seqA.seq[:20:] + seqB.seq[-20::]
print(seqC, '\n')

#d
seqD = seqA.seq[20:40] + seqC[11:31]
print(seqD, '\n')
```

# Sequence Analysis

```python
# a 
from Bio.SeqUtils import MeltingTemp

meltingTempSeqC = MeltingTemp.Tm_Wallace(seqC)
print(meltingTempSeqC, '\n')

# b harusnya from Bio.SeqUtils import GC aja
def GC(seq):
    from Bio.SeqUtils import gc_fraction
    return gc_fraction(seq) * 100

print(f'GC: {GC(seqC)}')
print(f'AT: {100-GC(seqC)}')
print('')

# c
from Bio.SeqUtils import molecular_weight
molecularWeightSeqC = molecular_weight(seqC)
print(molecularWeightSeqC)
print()

# d
from collections import Counter
print(Counter(seqC))
```

# DNA & mRNA Protein Synthesis

```python
mRNAB = seqB.seq.transcribe()
print(mRNAB, '\n')

protB = mRNAB.translate()
print(protB, '')
```

# Find Sequence Alignment & Similarities

```python
# a
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
localAlignment = pairwise2.align.localxx(seqC, seqD)
for i in localAlignment:
    print(format_alignment(*i))
```

```python
# b
def hamming(x, y):
    score = 0
    for i, j in zip(x, y):
        if i != j:
            score += 1
    return score

print(hamming(seqC, seqD))
print()

# c
from Levenshtein import distance
print(distance(seqC, seqD))
```

# plotting

```python
# Sequence C
import matplotlib.pyplot as plt
plt.title('Base nucleotide of Sequence C')
plt.bar(Counter(seqC).keys(), Counter(seqC).values())
plt.show()

plt.title('GC/AT Content of Sequence C')
plt.bar(['GC', 'AT'], [GC(seqC), 100-GC(seqC)])
plt.show()
```

```python
def dotplot(x, y):
    print(f' |{x}')
    for i in y:
        print(f'{i}|', end='')
        for j in x:
            if i == j:
                print('X', end='')
            else:
                print(' ', end='')
        print('')

dotplot(seqC, seqD)
```
