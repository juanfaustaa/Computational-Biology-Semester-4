Menghitung GC dan AT

```python
from Bio.Seq import Seq
import matplotlib.pyplot as plt
```

```python
Seq1 = Seq("GACTGACTCGGTACTGGATCGATCGATGGTACTGGATCAGTGAC")
```

```python
#GC content[Manual]
def findGC(sequence):
    G = sequence.count("G")
    C = sequence.count("C")
    total = len(sequence)
    percentage = (G+C)/total
    return percentage

gc1 = findGC(Seq1)
gc2 = 1 - gc1
print(f'GC Percentage: {gc1}')
print(f'AT Percentage: {gc2}')
```

```python
from Bio.SeqUtils import GC
print(GC(Seq1))
print(100-GC(Seq1))
```

Melting Temp

```python
from Bio.SeqUtils import MeltingTemp

#wallace
print(MeltingTemp.Tm_Wallace(Seq1))
#Nearest Neighbor
print(MeltingTemp.Tm_NN(Seq1))
#GC_content
print(MeltingTemp.Tm_GC(Seq1))
```

Molecular Weight

```python
from Bio.SeqUtils import molecular_weight
print(molecular_weight(Seq1))
```

Latihan Soal

```python
SeqA = Seq("AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC")
SeqB = Seq("CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG")
```

```python
print(f'Molecluar Weight of Sequence A: {molecular_weight(SeqA)}')
print(f'Molecluar Weight of Sequence B: {molecular_weight(SeqB)}')
```

```python
#Wallace Rule:
wallaceA = MeltingTemp.Tm_Wallace(SeqA)
wallaceB = MeltingTemp.Tm_Wallace(SeqB)
print(f'Melting Point Sequence A: {wallaceA}')
print(f'Melting Point Sequence B: {wallaceB}')
plt.title("Melting Temp Wallace")
plt.bar(["Melting Temp A", "Melting Temp B"], [wallaceA, wallaceB])
plt.show()
```

```python
#GC Content:
print(f'GC Count on Sequence A: {GC(SeqA)}%')
print(f'GC Count on Sequence B: {GC(SeqB)}%')
```

```python
#AT Content:
print(f'AT Count on Sequence A: {100-GC(SeqA)}%')
print(f'AT Count on Sequence B: {100-GC(SeqB)}%')
```

```python
plt.bar(["GC SeqA", "AT SeqA", "GC Seq B", "AT Seq B"], [GC(SeqA), 100-GC(SeqA), GC(SeqB), 100-GC(SeqB)])
```
