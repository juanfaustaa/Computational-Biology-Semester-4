```python
from Bio.Seq import Seq
```

```python
seqA = Seq('TACGATGCATGCATGCGGACGGACTAGGTACGACTACGT')
seqB = Seq('ACTACTATCATCATCACTATATATCTATACTCTAACTAT')
seqC = Seq('ATCTACGACGTAGCTACGTAGTCATCATAAGAACTAGCTGATGAGCC')
```

```python
from Bio import pairwise2
```

```python
localAlignment = pairwise2.align.localxx(seqA, seqB)
from Bio.pairwise2 import format_alignment
for i in localAlignment:
    print(format_alignment(*i))
```

```python
globalAlignment = pairwise2.align.globalxx(seqA, seqC)
for i in globalAlignment:
    print(format_alignment(*i))
```

```python
# hamming distance: berapa perbedaan dalam sequence
def hamming(x,y):
    score = 0
    for a, b in zip(x, y):
        if(a!=b):
            score+=1
    return score

print(hamming(seqA, seqB))
```

```python
#levenshtein distance: berapa kali suatu sequence harus diubah (substituion, insertion, or deletion) agar menjadi sequence yang dibandingkan.
from Levenshtein import distance
print(distance(seqA, seqC))
```

```python
def dotplot(x,y):
    print(f" | {y}")
    print("---------------")
    for i in x:
        print(f'{i}| ', end='')
        for j in y:
            if i == j:
                print('X', end='')
            else:
                print(' ', end = '')
        print('')
```

```python
dotplot(seqA, seqB)
```
