```python
from Bio import SeqIO
```

```python
#parse -> read multiple fasta file
#cara pertama, pake for loop
iterator = SeqIO.parse('sequence.fasta', 'fasta')
for i in iterator:
    print(i)
    
print('------------------------------------------------')
# cara kedua, dijadiin list, dan bisa akses per datanya

parse = list(SeqIO.parse('sequence.fasta', 'fasta'))
print(parse[0])

for i in range (0, len(parse)):
    print(parse[i])
    print('\n')
```

```python
#read -> read singular fasta file

read = SeqIO.read('sequence.fasta', 'fasta')

print('Name:', read.name)
print('ID:', read.id)
print('Description:', read.description)
print('Sequence:', read.seq)
```

```python
seqA = read.seq


print('Complement: ', seqA.complement())
print('Translation: ', seqA.translate())
```

```python
from Bio.PDB import PDBParser
```

```python
pdb_parser = PDBParser()
```

```python
structure_pdb = pdb_parser.get_structure('4ciw', '4ciw.pdb')
```

```python
model_pdb = structure_pdb[0]
```

```python
for c in model_pdb:
    print('Chain: ', c)
    for r in c:
        print('Residue: ', r)
        for a in r:
            print('Atom: ', a)
```
