```python
from Bio import SeqIO
```

```python
read = SeqIO.read('sequence.fasta', 'fasta')

sequence = read.seq

print(sequence)
```

```python
seqA = sequence[:round(len(sequence)/2)]
seqB = sequence[round(len(sequence)/2):]

print(seqA)
print(seqB)
```

```python
from Bio.SeqUtils import MeltingTemp

mtA = MeltingTemp.Tm_NN(seqA)
print('Melting Point Sequence A: ', mtA)
```

```python
from Bio.SeqUtils import GC

gcb = GC(seqB[:20])
atb = 100 - gcb

print('GC B: ', gcb)
print('AT B: ',atb)
```

```python
import matplotlib.pyplot as plt
```

```python
gc_a = GC(seqA)
gc_b = GC(seqB)

at_a = 100 - gc_a
at_b = 100 - gc_b
```

```python
plt.title('A and B GC Content')
plt.bar(['GC A', 'GC B'], [gc_a, gc_b])
plt.show()

plt.title('A and B AT Content')
plt.bar(['AT A', 'AT B'], [at_a, at_b])
plt.show()
```

```python
rna_b = seqB.transcribe()

aa_b = rna_b.translate()

print('Asam amino dari sequence B: ', aa_b)
```

```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
```

```python
first_10 = seqB[:10]
last_10 = seqB[-10:]
localalignment = pairwise2.align.localxx(first_10, last_10)

for i in localalignment:
    print(format_alignment(*i))
```
